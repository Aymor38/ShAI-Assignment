# **Descriptive Statistics**

```python
df.describe()
```

```python
basepay_range = df['BasePay'].max() - df['BasePay'].min()
totalpay_range = df['TotalPay'].max() - df['TotalPay'].min()
totalpayben_range = df['TotalPayBenefits'].max() - df['TotalPayBenefits'].min()
print("Base Pay Range: ", basepay_range)
print("Total Pay Range: ", totalpay_range)
print("Total Pay with Benefits Range: ", totalpayben_range)
```

# **Data Cleaning**

```python
df.drop(['Id','Notes', 'Status'], axis=1, inplace=True)
#remove the empty columns, no information gained from them, and so dropping null values doesn't empty out the whole dataset
#Id is not needed as the index is the same
df.head()
```

```python
df['BasePay'].fillna(df['BasePay'].mean(), inplace=True) # after removing emptying columns, fill missing values with mean value of their respective columns
df['OvertimePay'].fillna(df['OvertimePay'].mean(), inplace=True)
df['OtherPay'].fillna(df['OtherPay'].mean(), inplace=True)
df['Benefits'].fillna(df['Benefits'].mean(), inplace=True)

df.head()
```

```python
df.isnull().sum()
```

```python
df.to_csv('/content/Cleaned_Salaries.csv', index=False)
```

# **Basic Data Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalPay'])
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
job_counts = df['JobTitle'].value_counts()
top_jobs = job_counts.head(10)  # Considering top 10 job for better visualization
plt.figure(figsize=(10, 10))
plt.pie(top_jobs, labels=top_jobs.index, autopct='%1.1f%%', startangle=180)
plt.title('Proportion of Employees in Top 10 Jobs')
plt.show()
```

# **Grouped Analysis**

```python
df.groupby('JobTitle')['TotalPay'].agg(['mean', 'count', 'min', 'max', 'std']).sort_values(by='mean', ascending=False).head(10)
```

# **Simple Correlation Analysis**

```python
correlation_salary_basepay = df['TotalPay'].corr(df['BasePay'])
print("\nCorrelation between TotalPay and BasePay:", correlation_salary_basepay)

plt.figure(figsize=(8, 6))
sns.scatterplot(x='TotalPay', y='BasePay', data=df)
plt.title('Scatter Plot: TotalPay vs BasePay')
plt.xlabel('TotalPay')
plt.ylabel('BasePay')
plt.show()
```
